```python
import cmocean
import numpy as np
import netCDF4 as netcdf
import matplotlib.pyplot as plt
from kalpana.plotting import plot_nc
import cartopy.crs as ccrs
```

### Example of ADCIRC output plots

Maximum water level output (maxele.63.nc)

```python
## load the NetCDF dataset from a specified file path
nc = netcdf.Dataset(r'../../adds/inputs_examples/maxele.63.nc', 'r')
```

```python
## generate a plot of the maximum water levels
plot_nc(nc, 'zeta_max', levels = np.arange(0, 3.25, 0.25), background_map = True, cbar = True, cb_label = 'Max water level [mMSL]')
```

Same plot but defining the axis outside the function to have more control of the plot

```python
fig, ax = plt.subplots(figsize = (8,6), subplot_kw={'projection': ccrs.PlateCarree()}, 
                    constrained_layout=True)
plot_nc(nc, 'zeta_max', levels = np.arange(0, 3.25, 0.25), background_map = True, cbar = True, cb_label = 'Max water level [mMSL]',
        ax = ax, fig = fig)
ax.set_title('Maxele.63.nc')
```

Time-varying water level (fort.63.nc)

```python
nc = netcdf.Dataset(r'../../adds/inputs_examples/fort.63.nc', 'r')
nc2 = netcdf.Dataset(r'../../adds/inputs_examples/fort.74.nc', 'r')
```

```python
plot_nc(nc, 'zeta', levels = np.arange(-1.5, 1.6, 0.25), background_map = True, cbar = True, cb_label = 'Max water level [mMSL]', cmap = cmocean.cm.balance, ts = 800)
```

Time-varying water level with wind field (fort.63.nc and fort.74.nc)

```python
plot_nc(nc, 'zeta', levels = np.arange(-1.5, 1.6, 0.25), background_map = True, cbar = True, cb_label = 'Max water level [mMSL]', cmap = cmocean.cm.balance, ts = 800,
        ncvec = nc2, dxvec=0.05, dyvec=0.05, vecsc=700)
```
